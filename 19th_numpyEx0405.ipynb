```python
x1=np.array([1.0,2.0,3.0])
y1=np.array([5.0,10.0,15.0])
x2=np.array([[1.0,2.0],[3.0,4.0]])
y2=np.array([[5.0,10.0],[15.0,20.0]])

print("x1+y1 :")
print(x1+y1)
print()
print("x1*y1 :")
print(x1*y1)
print()
print("x2+y2 :")
print(x2+y2)
print()
print("x2*y2 :")
print(x2*y2)
```

```python
import matplotlib.pyplot as plt

from matplotlib.image import imread

x=np.arange(0,2*np.pi+0.1,0.1)
print("x")
print(x)
print()
y1=np.sin(x)
print("y1 :")
print(y1)
print()
y2=np.cos(x)
print("y2 :")
print(y2)
print("---------------------")
print("여기서부터 그래프")
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()#범례 나타냄
plt.show()
```

```python
g= np.array([2,3,6,1,4,5])
print("g=",g)
print("np.argmax(g)=",end=" ")
print(np.argmax(g))
print("g[np.argmax(g)]=",g[np.argmax(g)])#g[2]=6
print("---------------")
t=np.array([[0.1,0.8,0.2],[0.3,0.2,0.5],[0.9,0.5,0.3]])
print("t :")
print(t)
print()
print("np.argmax(t, axis=1)=")
print(np.argmax(t, axis=1))#0이면 컬럼에서 고름
#1이면 행에서 고름
print(np.argmax(t, axis=0))
#0이면 컬럼에서 고름
```
